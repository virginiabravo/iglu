## Problema 1: Clasificador de temperatura
Crea un programa que imprima los números desde -30 hasta 50.

Para cada número, además de imprimirlo, el programa debe mostrar un texto según estas reglas:

- si el número es menor que 0, mostrar "Muy frío"
- si el número está entre 0 y 24, mostrar "Templado"
- si el número es mayor o igual a 25, mostrar "Caluroso"

```
# Lista desde -30 al 50
lista_numeros = range(-30, 51)
print(lista_numeros)

# Iteramos sobre cada número en el rango especificado
for x in range(-30, 51):
  # Si el número es menor a 0, la temperatura es muy fría
  if(x < 0):
    print(x, "Muy frío")
  # Si el número está entre 0 y 24 (ambos inclusive), es templado
  elif (x >= 0 and x <= 24):
    print(x, "Templado")
  # Si no cumple ninguna de las anteriores (es 25 o mayor), es caluroso
  else:
    print(x, "Caluroso")
```

## Problema 2. Números especiales del robot
Crea un programa que imprima los números del 1 al 100 que sean divisibles por 5.

Además, el programa debe cumplir estas condiciones:
- si el número es divisible por 5, imprimir el número junto con el texto "Bip"
- si el número también es divisible por 10, imprimir el número junto con el texto "Bip Bip"

```
# Lista del 1 al 100
lista_robot = range(1, 101)

# Iteramos sobre cada número del 1 al 100
for x in range(1, 101):
  # Verificamos primero si es divisible por 5 y por 10 al mismo tiempo (el caso más restrictivo)
  if (x % 5 == 0 and x % 10 == 0):
    print(x, "Bip Bip")
  # Si no es divisible por 10, verificamos si al menos es divisible por 5
  elif (x % 5 == 0):
    print(x, "Bip")
```

## Problema 3. Detector de puntaje gamer
Crea un programa que solicite al usuario dos números, los sume y luego muestre un mensaje según el resultado.

Reglas:
- si la suma es menor que 50, mostrar "Nivel principiante"
- si la suma es mayor o igual a 50 y menor que 80, mostrar "Nivel - intermedio"
- si la suma es mayor o igual a 80, mostrar "Nivel pro"

```
# Solicitamos al usuario que ingrese dos puntajes y los convertimos a números enteros (int)
p1 = int(input("Ingresa primer puntaje: "))
p2 = int(input("Ingresa segundo puntaje: "))

# Calculamos la suma de ambos puntajes
suma = p1 + p2
print("") # Imprimimos un espacio en blanco para separar el resultado como en el ejemplo

# Evaluamos la suma para determinar el nivel del jugador
if (suma < 50):
  print("Resultado: Nivel principiante")
# Si la suma es mayor o igual a 50 pero menor a 80
elif (suma >= 50 and suma < 80):
  print("Resultado: Nivel intermedio")
# Si la suma es 80 o superior
else:
  print("Resultado: Nivel pro")
```

## Problema 4. Calculador de perfil musical
Crea un programa que evalúe el perfil del usuario según su edad y sus gustos.

El usuario debe recibir los siguientes parámetros:
- edad
- le_gusta_musica
- le_gusta_bailar

Y debe retornar uno de los siguientes mensajes:
- "Eres una estrella de la fiesta" si la persona tiene 18 o más años, - le gusta la música y le gusta bailar
- "Te gusta la música, pero no tanto bailar" si le gusta la música pero no le gusta bailar
"- Tienes tu propio estilo" en cualquier otro caso

```
# Solicitamos los datos del usuario: edad (entero) y sus preferencias (texto)
edad = int(input("Ingresa tu edad: "))
le_gusta_musica = input("¿Te gusta la música? (si/no): ")
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ")

# Comparamos las tres condiciones requeridas para ser "estrella de la fiesta"
if (edad >= 18 and le_gusta_musica == "si" and le_gusta_bailar == "si"):
  print("Eres una estrella de la fiesta")
# Si le gusta la música, pero respondió que no al baile
elif (le_gusta_musica == "si" and le_gusta_bailar == "no"):
  print("Te gusta la música, pero no tanto bailar")
# Cualquier otra combinación cae en esta categoría
else:
  print("Tienes tu propio estilo")
```
